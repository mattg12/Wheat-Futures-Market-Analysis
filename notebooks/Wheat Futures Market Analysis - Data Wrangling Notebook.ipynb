## Note on 'Rolling' Futures

When I say the data in 'continuous front month,' what that means is that for each day in the dataset, the prices recorded are for the (as of that day) 'front month' contract. Because futures expire, the front month contract changes periodically and predictably, and market participants need to 'roll' their exposure from one month to the next before the contract expires - meaning they close out their position in the front month contract and enter an offsetting position in the next contract, so that their 'position' is held even as the contract expires.

For analysis, it is not exactly appropriate to always use the 'front month' contract, since market participants 'roll' their exposure from the front to second month in advance of the front month expiration. The reason for this is that liquidity diminishes significantly as the contract nears expiration. As a result, prices for the 'front month' contract can have extreme moves during the final days of the contract. I have pulled in the continuous front and continuous second month contract time series from quandl.An appropriate data cleaning workflow would involve writing an algorithm to intelligently 'roll' between 1st and 2nd contracts (described in blog post linked below). In this project, I have not done this, as I was not able to get an explanation for how quandl rolls their futures in the free dataset, and fixing this would take significant time for potentially small value-add.  

Given that the downside to holding the front month future to expiry is that you are vulnerable to extreme single day moves, I think it is a reasonable assumption that the results I find in this study in fact provide conservative estimates, as in a live trading model, I would not hold the illiquid first contract to the last day.

## Returns and Volatility Information

Note: with the exception of the Parkinson Ratio, these formulas may not represent precisely how I calculated the metrics in my dataset, as I chose to rely to pandas' built in functionality. Any differences arising from differences in exact formulas should not affect analysis, but I wanted to provide formulas for the technical reader, to show the relationships between prices and the various metrics. Filtered volatility in particular is something I ended up not using, but am interested in exploring further, as it provides several new parameters to tune over.

Return:

$$x_t = Log\biggl(\frac{P_t}{P_t-1}\biggl)$$

Volatility:

$$\sigma_x ' = \sqrt{\frac{1}{n}\sum_{t=1}^{n} x_t^{2}}$$

Filtered Volatility:

$$\sigma = \sqrt{\frac{1}{\Omega}\sum_{t=1}^{n}\lambda ^{t}  x_t^{2}}$$
$$\Omega = \sum_{t=1}^{n}\lambda^{t} = (\lambda + \lambda^{2} + \lambda^{3} + ... + \lambda^{n}) \approx \frac{1}{1-\lambda}$$

Parkinson Ratio:

$$P = \sqrt{\frac{1}{(n)}\sum_{i=1}^{n}\frac{1}{4log2}\biggl[log\frac{S_H}{S_L}\biggl]^{2}}$$

Rule:

Compare, P = 1.67$\sigma '$. If P >> 1.67$\sigma '$, the market is more likely to be rangebound. If P << 1.67$\sigma '$, then the market is trending.


Source of returns and volatility formulas: Taleb, N.N. (1997). *Dynamic Hedging: Managing Vanilla and Exotic Options*, New York: John Wiley & Sons, Inc. 

## Data Preprocessing

Below I have written functions to preprocess my data, which includes handling null values and calculating some important descriptive statistice on the daily prices. In order to fill in null values, I used a simple algorithm to fill in missing prices with what I did have, on the assumption that prices recorded the same day would be 'close enough'. For example: if 'Last' was missing, I used 'Settle'; If 'High' was missing, I used the highest of the prices I had, etc. After filling in null values, I calculated rolling means, standard deviations, and Parkinson Ratios (another form of volatility described above) using a range of lookbacks.

```python
def parkinson_ratio(df, n):
    '''Given a timeseries of daily prices, return the n period parkinson ratio'''
    
    df['parkinson'] = (1/(4*np.log(2))) * ((np.log(df['High']/df['Low'])) ** 2)
    df['{}d_parkinson'.format(n)] = np.sqrt(df['parkinson'].rolling(n).mean())

    
def preprocess_data(df, ranges):
    '''Given timeseries data, generate columns of rolling averages and historical vol'''

    # Deal with missing data
    df['Last'].fillna(wheat['Settle'], inplace=True)
    df['Open'].fillna(wheat['Settle'].shift(-1), inplace=True)
    df['High'] = np.where(df['High'].isnull(), df[['Open','High','Low','Last','Settle']].max(axis=1), df['High'])
    df['Low'] = np.where(df['Low'].isnull(), df[['Open','High','Low','Last','Settle']].min(axis=1), df['Low'])

    # calculate exponentially weighted moving averages
    for r in ranges:
        df['{}dma'.format(r)] = df['Settle'].ewm(span = r).mean()
        
    # calculate forward looking moving averages
    for r in ranges:
        df['{}d_fma'.format(r)] = df['Settle'].iloc[::-1].ewm(span = r).mean()
    
    # calculate rolling volatility
    for r in ranges:
        df['{}d_vol'.format(r)] = df['Settle'].rolling(r).std()
    
    # calculate parkinson ratio
    for r in ranges:
        parkinson_ratio(df, r)
```

## Data Preprocessing: Add information about trends

The following functions come from extensive EDA I did in order to try to identify major trends and turning points in the wheat price history. In the context of data preparation, I need to use these functions to create categorical, and ultimately dummy variables for Regime (Up, Down, Flat, or Unclassified) and Turning Point (Top, Bottom, or Unclassified). I will export both a one-hot encoded df (for modeling) as well as one which retains categories (which I need for EDA).

### High-Level Overview

Below is a basic step by step guide to what the functions below are doing to prepare my data.

**Steps**
1. Perform a seasonal decomposition on the data to extract the underlying trend, after accounting for cyclicality and noise. Note that the choice of what frequency to use for the decomposition is an decision that needed to be made, and presents an opportunity for further refinement.
2. From the trend component of the seasonal decomposition, find the relative maxima and minima, or 'turning points' of the trend. Again, the order over which to search from relative extrema is a parameter that can be tuned.
3. Given the turning points of the trend, search around each one to find the actual extreme value (high or low) of the underlying time series.
4. Filter out the data points between turning points, and record each trend (period starting at one turning point and ending at the next), recording data such as total return, duration, start, end, end extreme values.
5. Further filter the trends data by re-assigning trends below a certain threshold for percent return as 'Flat.' Anything else is 'Unclassified.'

```python
def get_decomposition(df, start, stop, freq=60, column='Settle'):
    '''Return seasonal decomposition for given timeframe and frequency
    Plot the original series as well as the decomposition'''
    
    # run the seasonal decomposition
    series = wheat[start:stop][column]
    decomp = seasonal_decompose(series, freq=freq)
    
    # plot the decomposition
    fig, axes = plt.subplots(nrows=4, ncols=1, figsize = (15,15))
    series.plot(ax=axes[0], title='Price')
    decomp.trend.plot(ax=axes[1], title='Trend')
    decomp.seasonal.plot(ax=axes[2], title='Seasonal')
    decomp.resid.plot(ax=axes[3], title='Residual')
    plt.tight_layout();
    
    return decomp
```

```python
def find_trend_turning_points(df, start, stop, decomp, order=5):
    '''Returns the local maxima and minima of smoothed time series'''
    
    # for local maxima
    maxima = argrelextrema(decomp.trend.values, np.greater, order=order) # find the location of the maxes
    max_dates = [df[start:stop].iloc[i].index for i in maxima] # get the corresponding dates
    peaks = [date for date in df[start:stop].index if date in max_dates[0]]

    # for local minima
    minima = argrelextrema(decomp.trend.values, np.less, order=order)# find location of the mins
    min_dates = [df[start:stop].iloc[i].index for i in minima] # get the corresponding dates
    troughs = [date for date in df[start:stop].index if date in min_dates[0]]
    
    return peaks, troughs
```

```python
def find_exact_turning_points(df, peaks, troughs, order=10):
    '''Given a time series dataframe and dates of smoothed maxima and minima, find
    the exact high and low points by searching around the trend max/mins'''
    
    highs = []
    lows = []
    
    # find the exact maxs
    for date in peaks:
        start = date - pd.Timedelta(order, unit='D')
        stop = date + pd.Timedelta(order, unit='D')
        high_price_loc = df[start:stop]['High'].idxmax()
        highs.append(high_price_loc)
    
    # find the exact mins
    for date in troughs:
        start = date - pd.Timedelta(order, unit='D')
        stop = date + pd.Timedelta(order, unit='D')
        low_price_loc = df[start:stop]['Low'].idxmin()
        lows.append(low_price_loc)
        
    return highs, lows
```

```python
def is_top_or_bottom(date, highs, lows):
    '''Check if a given date is found in list of highs or lows'''
    
    if date in highs:
        classifier = 'Top'
    elif date in lows:
        classifier = 'Bottom'
    else:
        classifier = 'Unclassified'
        
    return classifier
```

```python
def filter_for_turns(df, highs, lows):
    '''Given a time series of prices and dates of highs and lows, filter the dataframe
    and return a new df which only contains turning points'''
    
    # Classify turning points
    df['Turning_Point'] = df.index
    df['Turning_Point'] = df['Turning_Point'].apply(is_top_or_bottom, args=(highs, lows))
    
    # Filter out points that aren't turns
    unclass = df['Turning_Point'] == 'Unclassified'
    endpoints = df[~unclass]
    
    return endpoints
```

```python
def trend_finder(df, threshold=10):
    '''Given a dataframe of turning points from a financial time series, return a dataframe summarizing
    the trends in the series'''
    
    # Initialize an empty list to contain trends
    trends = []

    # Initialize trackers to indicate tops and bottoms have been found
    top = False
    bottom = False

    # Loop through df, collecting pairs of tops and bottoms
    # each pair is a single trend
    for i, row in df.iterrows():
        if row['Turning_Point'] == 'Top':
            if top == False:
                peak_date = i
                peak_price = row['High']
                top = True
            elif top == True:
                if row['High'] > peak_price:
                    peak_date = i
                    peak_price = row['High']
                    top = True
                else:
                    continue
        if row['Turning_Point'] == 'Bottom':
            if bottom == False:
                trough_date = i
                trough_price = row['Low']
                bottom = True
            elif bottom == True:
                if row['Low'] < trough_price:
                    trough_date = i
                    trough_price = row['Low']
                    bottom = True
                else:
                    continue
        
        # Once a top and bottom have been paired, save the dates and prices as a dict
        if top == True & bottom == True:
            trend = {'High_Date': peak_date,
                     'High_Price': peak_price,
                     'Low_Date': trough_date,
                     'Low_Price': trough_price}
            trends.append(trend)
    
            if peak_date > trough_date:
                bottom = False
            else:
                top = False
    
    # Create a dataframe to summarize each trend
    trends = pd.DataFrame(data=trends)
    trends['Direction'] = np.where(trends['High_Date'] > trends['Low_Date'],'Up','Down')
    trends['Start_Date'] = trends[['High_Date', 'Low_Date']].min(axis=1)
    trends['End_Date'] = trends[['High_Date', 'Low_Date']].max(axis=1)
    trends['Duration'] = np.where(trends['Direction'] == 'Up',
                        trends['High_Date'] - trends['Low_Date'],
                        trends['Low_Date'] - trends['High_Date'])
    
    trends['Return'] = np.where(trends['Direction'] == 'Up',
                        ((trends['High_Price'] - trends['Low_Price']) / trends['Low_Price']) * 100,
                        ((trends['Low_Price'] - trends['High_Price']) / trends['High_Price']) * 100)
    
    # Classify periods below the threshold as flat
    trends.loc[abs(trends['Return']) < threshold, 'Direction'] = 'Flat'
    
    # Code to handle one corner-case that came up in the data
    for i, row in trends.iterrows():
        if (row['Direction'] == 'Up') and (row['High_Price'] < row['Low_Price']):
            trends.loc[i, 'Direction'] = 'Down'
            trends.loc[i, 'High_Price'] = row['Low_Price']
            trends.loc[i, 'Low_Price'] = row['High_Price']
            trends.loc[i, 'High_Date'] = row['Low_Date']
            trends.loc[i, 'Low_Date'] = row['Low_Price']
    
    # Aggregate flat periods
    is_flat = False

    for i, row in trends.iterrows():
        if row['Direction'] == 'Flat':
            if is_flat == False:
                start_date = row['Start_Date']
                is_flat = True
            elif is_flat == True:
                trends.loc[i,'Start_Date'] = start_date
        if row['Direction'] != 'Flat':
            is_flat = False
            continue
    
    trends = trends.groupby(by=['Start_Date', 'Direction']).agg({'High_Price': max, 'Low_Price': min,
                                                             'Duration': sum, 'Return': 'mean',
                                                             'End_Date': max,'High_Date': max, 'Low_Date': max})[['High_Price',
                                                                'Low_Price', 'Duration', 'Return', 'End_Date',
                                                                'High_Date', 'Low_Date']].reset_index()
        
    return trends
```

In the cells below, I run the above functions to extract information about trends, which I use below to add labels to my original dataframe indicating which 'Regime' (Up, Down, Flat, or Unclassified) each observation falls under.

```python
df = wheat
start = wheat.index[0]
stop = wheat.index[-1]

decomp = get_decomposition(df, start, stop)
peaks, troughs = find_trend_turning_points(df, start, stop, decomp)
highs, lows = find_exact_turning_points(df, peaks, troughs)
filtered_df = filter_for_turns(df, highs, lows)
trends = trend_finder(filtered_df)
```

```python
def create_date_labels(trends):
    '''Create a dataframe with dates and regime labels'''
    
    # Split trends df into separate dataframes for each label
    uptrends = trends[trends['Direction'] == 'Up']
    downtrends = trends[trends['Direction'] == 'Down']
    ranges = trends[trends['Direction'] == 'Flat']
    
    # For uptrends create a df of all dates included, with the label 'Up'
    uptrends.reset_index(inplace=True)
    up_dates = []
    for i, row in uptrends.iterrows():
        uptrend_dates = [date for date in pd.date_range(start=row['Low_Date'], end=row['High_Date'])]
        up_dates.extend(uptrend_dates)
    ups = pd.DataFrame(data=['Up' for date in up_dates], index=up_dates)
    
    # Do the same for downtrends, with the label 'Down'
    downtrends.reset_index(inplace=True)
    down_dates = []
    for i, row in downtrends.iterrows():
        downtrend_dates = [date for date in pd.date_range(start=row['High_Date'], end=row['Low_Date'])]
        down_dates.extend(downtrend_dates)
    downs = pd.DataFrame(data=['Down' for date in down_dates], index=down_dates)
    
    # Same for ranges, being careful to track which date is start/end
    ranges.reset_index(inplace=True)
    flat_dates = []
    for i, row in ranges.iterrows():
        if row['High_Date'] > row['Low_Date']:
            range_dates = [date for date in pd.date_range(start=row['Low_Date'], end=row['High_Date'])]
        elif row['Low_Date'] > row['High_Date']:
            range_dates = [date for date in pd.date_range(start=row['High_Date'], end=row['Low_Date'])]
        flat_dates.extend(range_dates)
    ranges = pd.DataFrame(data=['Flat' for date in flat_dates], index=flat_dates)
    
    # Rename the labels column to a common name: 'Regime'
    ups.columns = ['Regime']
    downs.columns = ['Regime']
    ranges.columns = ['Regime']
    
    # Merge all into one df
    labels = pd.concat([ranges, ups, downs], sort=True)
    labels.index.name = 'Date'
    labels['Date'] = labels.index
    labels.drop_duplicates(subset='Date', keep=False, inplace=True)
    labels.drop(columns='Date', inplace=True)
    labels.sort_index(inplace=True)
    
    return labels
```

```python
labels = create_date_labels(trends)
```

```python
wheat = wheat.merge(labels, on='Date', how='left')
```
